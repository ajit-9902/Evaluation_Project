Lets See the skewness in the datasets 

```python
x=baseball.drop('Wins',axis=1)
x.shape
```

```python
for columns in x:
    sns.distplot(x[columns],kde=True)
    plt.xlabel(columns)
    plt.ylabel('Count')
    plt.show()
```

```python
x.skew()
```

 For skewness values between -0.5 and 0.5, the data exhibit approximate symmetry. Skewness values within the range of -1 and -0.5 (negative skewed) or 0.5 and 1(positive skewed) indicate slightly skewed data distributions

Hence from the above data we see that saves and errors are highly skewed 

```python
#Applying Log Transformation to remove the skewness
columns=['Saves','Error']
for i in columns:
    x[i]=np.abs(np.log(x[i]))
```

```python
x.skew()
```

As we see the above table of skewness of saves and errors are removed.

We will standardize the datasets using Min-Max scaler since the dataset is also very small()

```python
minmax=MinMaxScaler()
x_minmax=minmax.fit_transform(x)
x_minmax
```

This estimator scales and translates each feature individually such that it is in the given range on the training set between 0 and 1

```python
x_scaled=pd.DataFrame(x_minmax,columns=x.columns)
```

```python
x_scaled
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(x_scaled.corr() , annot = True , fmt = '.2%' , cmap = 'Spectral')
```

PCA will take all the columns in the dataset which are highly correlated and reduce the dimensionality of the datasets.

```python
bestaccuracy=0
for i in range(45):
    X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=i)
    lg=LinearRegression()
    lg.fit(X_train,Y_train)
    pred=lg.predict(X_test)
    accuracy=r2_score(Y_test,pred)
    if accuracy>bestaccuracy:
        bestaccuracy=accuracy
        randomState=i
```
